```python
#useful functions
# each name will come is as x, as it comes in its split and we pick position 0

def firstname(x):
    return x.split()[0]
```

```python
firstname("Wale Olajumoke")
```

```python
firstname("Elon Musk")
```

```python
df["firstname"] = df["Name"].apply(firstname)
df
```

```python
def lastname(x):
    return x.split()[1]
```

```python
lastname("Wale Olajumoke")
```

```python
lastname("Elon Musk")
```

```python
df["Last name"] = df["Name"].apply(lastname)
df
```
